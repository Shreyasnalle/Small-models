```python
survived_df = train_dataset[train_dataset["Survived"] == 1]
survived_df
```

```python
total_people = train_dataset.shape[0]
print(f"{survived_df.shape[0]} people survived out of {total_people}")
```

```python
total_male = (train_dataset["Sex"] == "male").sum()
print(f"Total number of male are {total_male}")
total_female = (train_dataset["Sex"] == "female").sum()
print(f"Total number of female are {total_female}")
```

```python
male_survived_df = survived_df[survived_df["Sex"] == "male"]
male_survived_df
```

```python
female_survived_df = survived_df[survived_df["Sex"] == "female"]
female_survived_df
```

```python
print(f"{male_survived_df.shape[0]} male survived out of {total_male}")
print(f"{(total_male) - (male_survived_df.shape[0])} male died")
print("")
print(f"{female_survived_df.shape[0]} female survived out of {total_female}")
print(f"{(total_female) - (female_survived_df.shape[0])} female died")
```

```python
total_upper_class_male = ((train_dataset["Pclass"] == 1) & (train_dataset["Sex"] == "male")).sum()
print(f"Total number of upper class men are {total_upper_class_male}")
total_middle_class_male = ((train_dataset["Pclass"] == 2) & (train_dataset["Sex"] == "male")).sum()
print(f"Total number of middle class men are {total_middle_class_male}")
total_lower_class_male = ((train_dataset["Pclass"] == 3) & (train_dataset["Sex"] == "male")).sum()
print(f"Total number of lower class men are {total_lower_class_male}")
print("")
total_upper_class_female = ((train_dataset["Pclass"] == 1) & (train_dataset["Sex"] == "female")).sum()
print(f"Total number of upper class men are {total_upper_class_female}")
total_middle_class_female = ((train_dataset["Pclass"] == 2) & (train_dataset["Sex"] == "female")).sum()
print(f"Total number of middle class men are {total_middle_class_female}")
total_lower_class_female = ((train_dataset["Pclass"] == 3) & (train_dataset["Sex"] == "female")).sum()
print(f"Total number of lower class men are {total_lower_class_female}")
```

```python
total_survived_upper_class_male = (male_survived_df["Pclass"] == 1).sum()
print(f"The total number of upper class male survived are {total_survived_upper_class_male} out of {total_upper_class_male}")
total_survived_middle_class_male = (male_survived_df["Pclass"] == 2).sum()
print(f"The total number of middle class male survived are {total_survived_middle_class_male} out of {total_middle_class_male}")
total_survived_lower_class_male = (male_survived_df["Pclass"] == 3).sum()
print(f"The total number of lower class male survived are {total_survived_lower_class_male} out of {total_lower_class_male}")
print("")
total_survived_upper_class_female = (female_survived_df["Pclass"] == 1).sum()
print(f"The total number of upper class female survived are {total_survived_upper_class_female} out of {total_upper_class_female}")
total_survived_middle_class_female = (female_survived_df["Pclass"] == 2).sum()
print(f"The total number of middle class female survived are {total_survived_middle_class_female} out of {total_middle_class_female}")
total_survived_lower_class_female = (female_survived_df["Pclass"] == 3).sum()
print(f"The total number of lower class female survived are {total_survived_lower_class_female} out of {total_lower_class_female}")
```

```python
total_male_survived = male_survived_df.shape[0]
percentage_of_male_survived  = (total_male_survived / total_male) * 100
print(f"The percentage of male survived is {percentage_of_male_survived:.4f}")
print("")
total_female_survived = female_survived_df.shape[0]
percentage_of_female_survived = (total_female_survived / total_female) * 100
print(f"The percentage of female survived is {percentage_of_female_survived:.4f}")
print("")
survival_mean_by_sex = train_dataset.groupby("Sex")["Survived"].mean()
print(type(survival_mean_by_sex))
print(f"The mean of male survival is {survival_mean_by_sex['male']:.4f}")
print(f"The mean of the female survival is {survival_mean_by_sex['female']:.4f}")
```

```python
print(f"Thus clearly showing that the female survival rate was much greater than the males")
```

```python
class_wise_survival = train_dataset.groupby("Pclass")["Survived"].mean()
total_upper_class_percentage_survival = class_wise_survival[1] * 100
total_middle_class_percentage_survival = class_wise_survival[2] * 100
total_lower_class_percentage_survival = class_wise_survival[3] * 100
print(f"Upper class survival percentage is {total_upper_class_percentage_survival:.4f}")
print(f"Middle class survival percentage is {total_middle_class_percentage_survival:.4f}")
print(f"Lower class survival percentage is {total_lower_class_percentage_survival:.4f}")
```

```python
print(f"Thus showing that upper class people were given priority while evacuating")
```

```python
class_sex_wise_survival = train_dataset.groupby(["Pclass", "Sex"])["Survived"].mean().mul(100).reset_index(name = "Survival_Percentage")
class_sex_wise_survival["Survival_Percentage"] = class_sex_wise_survival["Survival_Percentage"].round(4)
class_sex_wise_survival
```

```python
print(type(class_sex_wise_survival))
print("Class wise and sex wise data analaysis done, class wise gender based evaculation was done")
```

```python
survived_df_below18 = survived_df[survived_df["Age"] <= 18]
survived_df_below18
```

```python
below18_survived_percentage = ((survived_df_below18.shape[0]) / ((train_dataset["Age"] <= 18).sum())) * 100
print(f"{survived_df_below18.shape[0]} number of people below the age of 18 survived and the survival percentage of them is {below18_survived_percentage:.4f}")
```

```python
age18toage50_suvived_df = survived_df[(survived_df["Age"] > 18) & (survived_df["Age"] <= 50)]
age18toage50_suvived_df
```

```python
age18toage50_survived_percentage = ( age18toage50_suvived_df.shape[0] / (((train_dataset["Age"] > 18) & (train_dataset["Age"] <= 50)).sum()) ) * 100
print(f"{age18toage50_suvived_df.shape[0]} number of people in between the age of 18 and 50 survived and the survival percentage of them is {age18toage50_survived_percentage:.4f}")
```

```python
ageabove50_survived_df = survived_df[survived_df["Age"] >= 50]
ageabove50_survived_df
```

```python
ageabove50_survived_percentage = ((ageabove50_survived_df.shape[0]) / ((train_dataset["Age"] >= 50).sum())) * 100 
print(f"{ageabove50_survived_df.shape[0]} number of people above the age of 50 survived and the survival percentage of them is {ageabove50_survived_percentage:.4f}")
```

```python
print(f"Evacuation was done quite prominently when it comes to age")
```
