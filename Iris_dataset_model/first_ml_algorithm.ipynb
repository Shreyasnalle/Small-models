```python
from sklearn import tree
```

```python
# step 1, this is the collection of the data
features = [[150, 0], [170, 0], [130, 1], [120, 1]]
labels = ["apple", "apple", "orange", "orange"]
```

```python
# step 2, this is the training of the model
clf = tree.DecisionTreeClassifier()
clf = clf.fit(features, labels)
```

```python
# step 3, predicting the result
prediction = clf.predict([[120, 0]])
print(f"The fruit is {prediction[0]}")
```
